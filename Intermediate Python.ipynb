```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm

countries = ['Швейцария', 'Лихтенштейн', 'Канада', 'Украина', 'Беларусь', 'Мальта', 'Эстония', 'Австрия']
bith_koef = [10.5, 10.4, 10.2, 10.1, 10.0, 10.0, 9.9, 9.5]

#alpha - прозрачность от о (невидимый) до 1 (полностью видимый)
#facelolor - цвет отображемый значений на графике. Можно использовать ключевые слова либо RGB
#marker - маркер отображаемых значений на графике
#Типы маркеров можно рассмотреть по ссылке http://bit.do/marrrkers
#s = размер маркеров в пикселях. Чем выше значение, тем больше маркер. 
plt.scatter(countries, bith_koef, marker = "D", facecolor ='#D42CEE', alpha = 0.75, s = 75, cmap = 'hot')

#Создание заголовка графика
plt.title(' Birth coefficient')

#Создание заголовка оси х
plt.xlabel(' Countries')

#Создание заголовка оси y
plt.ylabel(' Сoefficient')

#Показ графика
plt.show()
```

```python
data = {'a': np.arange(50),
'c': np.random.randint(0, 50, 50),
'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator

# generate example data
import numpy as np
x,y = np.meshgrid(np.linspace(-1,1,15),np.linspace(-1,1,15))
z = np.cos(x*np.pi)*np.sin(y*np.pi)

# actual plotting example
fig = plt.figure()
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(x,y,z,rstride=1,cstride=1,cmap='viridis')
ax2 = fig.add_subplot(122)
cf = ax2.contourf(x,y,z,51,vmin=-1,vmax=1,cmap='viridis')
cbar = fig.colorbar(cf)
cbar.locator = LinearLocator(numticks=11)
cbar.update_ticks()
for ax in {ax1, ax2}:
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$y$')
    ax.set_xlim([-1,1])
    ax.set_ylim([-1,1])
    ax.set_aspect('equal')

ax1.set_zlim([-1,1])
ax1.set_zlabel(r'$\cos(\pi x) \sin(\p    i y)$')

plt.show()
```
